## Data analysis

```python
#.5% of total dataset sample
Sample_df =  df_tracks.sample(int(0.005*len(df_tracks)))

len(Sample_df)
```

```python
cm = df_tracks.drop(['key','mode'], axis=1).corr(method = 'pearson')
plt.figure(figsize=(14,6))
map = sns.heatmap(cm, annot = True, fmt = '.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='Black')
map.set_title('Correlation Heatmap between Variable')
map.set_xticklabels(map.get_xticklabels(), rotation=90)
```

```python
#Regression plot between loudness and energy
plt.figure(figsize = (10,8))
sns.regplot(data  =  Sample_df , y = "popularity" , x = "loudness" , color = 'c').set_title("Loudness vs popularity Correlation")
```

There is a clear, high correlation between loudness and energy of a song.

```python
#Regression plot between popularity and acoustics
plt.figure(figsize = (8,6))
sns.regplot(data  =  Sample_df , y = "popularity" , x = "acousticness" , color = 'b').set_title("Popularity  vs Acousticness Correlation")
```

There exist an inversely proportional relationship between popularity and Acousticness of a song.

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
popular = df_tracks.sort_values('popularity', ascending=False).head(10)
sns.barplot(y = 'genre', x = 'popularity', data = popular).set(title='Top 5 Genres by Popularity')
```

The popularity of the song depends on the genre as well. Dance is the most popular genre.

```python
#energy
plt.figure(figsize = (10,8))
sns.regplot(data  =  Sample_df , y = "popularity" , x = "energy" , color = 'c').set_title("energy vs popularity Correlation")
```

The greater the energy of the song the more the popularity

```python
#danceability
plt.figure(figsize = (10,8))
sns.regplot(data  =  Sample_df , y = "popularity" , x = "danceability" , color = 'c').set_title("Danceabiblity vs popularity Correlation")
```

The higher the danceabilty of the song the more popular it is.
